```python
from gpy_test.contour import contour
import numpy as np
import matplotlib.pyplot as plt
from gpy_test.config.contour import ContourConfig
```

```python
def show_contour(type_: str) -> None:
    contour_config = ContourConfig(imag_height=0.1, real_slack=1, type_=type_)

    f, df = contour(contour_config, eig_range=(0, 1))

    dts = np.linspace(0, 1, 10)
    zs = [f(t) for t in dts]
    dzs = [df(t) for t in dts]
    plt.plot(np.real(zs), np.imag(zs))

    # also plot the dz vectors at some points. Draw them as little arrows, starting at zs[t] and ending at zs[t] + dzs[t]
    for i in range(0, len(zs), 2):
        v = zs[i]
        dv = dzs[i] / 5  # rescale to make the plot look better
        plt.arrow(
            np.real(v),
            np.imag(v),
            np.real(dv),
            np.imag(dv),
            head_width=0.05,
            head_length=0.1,
            fc="k",
            ec="k",
        )
```

# Ellipse

```python
show_contour("ellipse")
```

# Circle

```python
show_contour("circle")
```

# Rectangle

```python
show_contour("rectangle")
```
