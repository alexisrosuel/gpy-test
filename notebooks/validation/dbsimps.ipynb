```python
from scipy.integrate import simpson, dblquad, quad
import numpy as np
from gpy_test.utils import dblsimpson
```

# 1D Simpson (complex)

```python
f = lambda x: np.exp(1j * x)

x = np.linspace(0, 2 * np.pi, 100)
y = f(x)

simpson(y=y, x=x)
```

# 2D Simpson (real)

```python
f = lambda x, y: x**2 + y**4

x_range = np.linspace(0, 2, 100)
y_range = np.linspace(0, 1, 50)
func_values = np.array([[f(x, y) for y in y_range] for x in x_range])

dblsimpson(func_values, x_range, y_range)
```

```python
dblquad(
    f, 0, 1, lambda x: 0, lambda x: 2
)  # warning: dlquad computes the integral of f(y, x) instead of f(x, y) as in dblsimpson!!
```

# 2D Simpson (complex)

```python
f = lambda x, y: np.exp(-1j * (x**2 + y**2))

x_range = np.linspace(-2, 2, 100)
y_range = np.linspace(-1, 1, 50)
func_values = np.array([[f(x, y) for y in y_range] for x in x_range])

dblsimpson(func_values, x_range, y_range)
```

```python
dblquad(
    lambda x, y: np.real(f(x, y)), -1, 1, lambda x: -2, lambda x: 2
)  # warning: dlquad computes the integral of f(y, x) instead of f(x, y) as in dblsimpson!!
```

```python
dblquad(
    lambda x, y: np.imag(f(x, y)), -1, 1, lambda x: -2, lambda x: 2
)  # warning: dlquad computes the integral of f(y, x) instead of f(x, y) as in dblsimpson!!
```
