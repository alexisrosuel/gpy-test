```python
from gpy_test.contour import _ellipse, _rectangle, _circle, contour
import numpy as np
import matplotlib.pyplot as plt
from gpy_test.config.contour import ContourConfig
from scipy.integrate import simpson, quad
```

```python
f = lambda z: np.exp(z)
t_range = np.linspace(0, 1, 10000)

# integrate of the circle, rectangle and ellipse contours
# in the complex plane

# circle
z, dz = _circle(-1, 1)
simpson(y=[f(z(t)) * dz(t) for t in t_range], x=t_range)
```

```python
%%timeit
quad(lambda t: np.real(f(z(t)) * dz(t)), 0, 1), quad(
    lambda t: np.imag(f(z(t)) * dz(t)), 0, 1
)
```

```python
%%timeit
quad(lambda t: f(z(t)) * dz(t), 0, 1, complex_func=True)
```
